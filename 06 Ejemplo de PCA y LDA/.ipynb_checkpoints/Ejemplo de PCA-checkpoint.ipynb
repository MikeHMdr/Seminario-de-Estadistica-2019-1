# Principal Components Analysis (PCA)

```python
# Bibliotecas a utilizar 
import numpy                   as     np                 # Uso de álgebra lineal, funciones vectoriales
import pandas                  as     pd                 # Trabajar con DataFrames
import matplotlib.pyplot       as     plt                # Para realizar gráficas
import seaborn                 as     sns                # Otras funciones para graficar

from sklearn.decomposition import PCA                    # Funcion de Análisis de Componentes Principales
sns.set()

import warnings
warnings.filterwarnings('ignore')
```

```python
# Creamos una tabla con puntos
np.random.seed(1)
matriz_diseño = np.random.multivariate_normal( np.array( [0,0] ), np.array( [[40.0, 45.0], [45.0, 55.0]] ), 100)
datos = pd.DataFrame( data = {'X':matriz_diseño[:,0], 'Y':matriz_diseño[:,1], 'Z':np.repeat('#9467BD',100) } )
datos.to_csv("datos_bivariados1.csv",index=False)

np.random.seed(1)
matriz_diseño = np.random.multivariate_normal( np.array( [0,0] ), np.array( [[40.0, 22.0], [22.0, 55.0]] ), 1000)
datos = pd.DataFrame( data = {'X':matriz_diseño[:,0], 'Y':matriz_diseño[:,1], 'Z':np.array( [ '#1F77B4' if x < 0 else '#D62728' for x in matriz_diseño[:,0] ] ) } )
datos.to_csv("datos_bivariados2.csv",index=False)

np.random.seed(1)
matriz_diseño = np.random.multivariate_normal( np.array( [0,0] ), np.array( [[40.0, 22.0], [22.0, 55.0]] ), 1000)
datos = pd.DataFrame( data = {'X':matriz_diseño[:,0], 'Y':matriz_diseño[:,1], 'Z':np.array( [ '#1F77B4' if x == True else '#D62728' for x in matriz_diseño[:,1] > -matriz_diseño[:,0] ] ) } )
datos.to_csv("datos_bivariados3.csv",index=False)
```

```python
# Leemos los datos previamente generados
datos = pd.read_csv("datos_bivariados1.csv")   # Datos con una sola clase
#datos = pd.read_csv("datos_bivariados2.csv")  # Datos con dos clases

matriz_diseño = np.array( [ np.array( datos["X"] ), np.array( datos["Y"] ) ] ).T
matriz_diseño[0:10,:]

datos.head()
```

```python
# Graficamos nuestros puntos
plt.figure( figsize=(16,6) )
plt.title("Nube de puntos")
plt.scatter( x=datos["X"], y=datos["Y"], c=datos["Z"], label="Puntos")
plt.axis('equal')
plt.show()
```

```python
# Mostramos el rango de ambas proyecciones sobre los ejes
x = np.round( datos["X"], 1 )
y = np.round( datos["Y"], 1 )
print( "Min(X)=", np.min(x), "Max(X)=", np.max(x), "Range(X)=", np.ptp(x) )
print( "Min(Y)=", np.min(y), "Max(Y)=", np.max(y), "Range(Y)=", np.ptp(y) )

# Histograma por colores
#sns.FacetGrid(datos, hue="Z", height=5, aspect=3, palette=["#1F77B4","#D62728"]).map(plt.hist,"X")
#plt.show()

# Graficamos nuestros puntos
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(16,4) )
ax1.set_title("Proyección sobre X")
ax1.hist( x=x, color="Orange" )
ax2.set_title("Proyección sobre Y")
ax2.hist( x=y, color="green"  )
plt.show()
```

```python
# Creamos nuestro constructor de clase para los componentes principales
modelo_pca = PCA(n_components=2).fit( matriz_diseño )
modelo_pca
```

```python
# Mostramos la varianza de cada componente principal y lo que representa de la varianza total
print(" Varianza Explicada:     ", modelo_pca.explained_variance_ )
print(" Varianza Explicada (%): ", modelo_pca.explained_variance_ratio_ )
print( 69.45781286 / (69.45781286 + 1.6540289) )
print(   1.6540289 / (69.45781286 + 1.6540289) )
```

```python
# Obtenemos nuestros componentes principales
# Estos ya vienen ordenados por la varianza explicada, es decir, arriba el de mayor varianza
componentes_principales = modelo_pca.components_
print( "Componentes Principales" )
print( componentes_principales )

# Por construcción, los componentes principales son ortogonales entre si
# Podemos validar que estos vectores son orgononales si el producto escalar/interior/punto es igual a cero
print( "Método 1" )
print( componentes_principales[0,:] @ componentes_principales[1,:] )
print( "Método 2" )
print( sum( componentes_principales[0,:] * componentes_principales[1,:] ) )
print( "Método 3" )
print( np.dot( componentes_principales[0,:], componentes_principales[1,:] ) )
```

```python
# Definimos una función que nos de las coordenadas (x,y) de nuestros componentes para poder graficarlos
def coordenadas_pc( componente_principal, valores_x, x0 = 0.0, y0 = 0.0 ):
    x1, y1 = componente_principal
    x      = np.array( [ np.min( valores_x ), np.max( valores_x ) ] ) 
    y      = y1 / x1 * x - y1 / x1 * x0 + y0
    return x,y
```

```python
# Graficamos los componentes principales y nuestros datos
# Vease que seguimos usando nuestros ejes iniciales para graficar
x1, y1 = coordenadas_pc( componentes_principales[0,:], datos["X"] )
x2, y2 = coordenadas_pc( componentes_principales[1,:], datos["X"] )
plt.figure( figsize=(16,6) )
plt.title("Nube de puntos")
plt.scatter( x = datos["X"], y = datos["Y"], c = datos["Z"], label="Puntos" )
plt.plot( x1, y1, c = "orange", label = "PC 1", linewidth=3, linestyle="--")
plt.plot( x2, y2, c = "green",  label = "PC 2", linewidth=2, linestyle="--")
plt.axis('equal')
plt.xlabel("Eje X")                             
plt.ylabel("Eje Y")
plt.show()
```

```python
nuevos_ejes = modelo_pca.fit_transform( matriz_diseño )

# Graficamos los componentes principales y nuestros datos
# Vease que seguimos usando nuestros ejes iniciales para graficar
x1, y1 = coordenadas_pc( componentes_principales[0,:], datos["X"] )
x2, y2 = coordenadas_pc( componentes_principales[1,:], datos["X"] )
plt.figure( figsize=(16,6) )
plt.title("Nube de puntos")
plt.scatter( x = nuevos_ejes[:,0], y = nuevos_ejes[:,1], c = datos["Z"], label="Puntos" )
plt.plot( x1, y1, c = "orange", label = "PC 1", linewidth=3, linestyle="--")
plt.plot( x2, y2, c = "green",  label = "PC 2", linewidth=2, linestyle="--")
plt.axis('equal')
plt.xlabel("Eje X")                             
plt.ylabel("Eje Y")
plt.show()
```

```python
x = np.round( nuevos_ejes[:,0], 1 )
y = np.round( nuevos_ejes[:,1], 1 )
print( "Min(X)=", np.min(x), "Max(X)=", np.max(x), "Range(X)=", np.ptp(x) )
print( "Min(Y)=", np.min(y), "Max(Y)=", np.max(y), "Range(Y)=", np.ptp(y) )

# Graficamos nuestros puntos
plt.figure( figsize=(16,4) )
plt.subplot(2,1,1)
plt.title("Proyección sobre PC 1")
plt.hist( x=x, color="Orange" )
plt.subplot(2,1,2)
plt.title("Proyección sobre PC 2")
plt.hist( x=y, color="green"  )
plt.show()

#datos_trans = pd.DataFrame( {"X":x, "Y":y,"Z":datos["Z"]} )
#sns.FacetGrid(datos_trans, hue="Z", height=5, aspect=3, palette=["#1F77B4","#D62728"]).map(plt.hist,"X")
#plt.show()

# Graficamos nuestros puntos
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(16,4) )
ax1.set_title("Proyección sobre PC 1")
ax1.hist( x=x, color="Orange" )
ax2.set_title("Proyección sobre PC 2")
ax2.hist( x=y, color="green"  )
plt.show()
```
