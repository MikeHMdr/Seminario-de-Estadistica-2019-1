# Linear Discriminant Analysis (LDA)

```python
# Bibliotecas a utilizar 
import numpy                   as     np                 # Uso de álgebra lineal, funciones vectoriales
import pandas                  as     pd                 # Trabajar con DataFrames
import matplotlib.pyplot       as     plt                # Para realizar gráficas
import seaborn                 as     sns                # Otras funciones para graficar

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics               import accuracy_score     # Función para calcular la precisión de clasificación
from sklearn.metrics               import confusion_matrix   # Función para calcular una matriz de confusión
from sklearn.model_selection       import KFold              # Función para realizar K-Fold Cross Validation
from sklearn.model_selection       import cross_val_score    # Función para realizar K-Fold-Cross Validation 

# https://sebastianraschka.com/Articles/2014_python_lda.html
# https://statquest.org/
# https://www.youtube.com/playlist?list=PLblh5JKOoLUIcdlgu78MnlATeyx4cEVeR

sns.set()

import warnings
warnings.filterwarnings('ignore')
```

## Datos :: Simulados

```python
# Leemos datos de ejemplo
# Estos fueron creados en el Archivo de 'Ejemplo de PCA'
#datos = pd.read_csv("datos_bivariados2.csv")   # Datos con dos clases
datos = pd.read_csv("datos_bivariados3.csv")   # Datos con dos clases
datos.head()
```

```python
# Graficamos nuestros puntos
plt.figure( figsize=(16,6) )
plt.title("Nube de puntos")
plt.scatter( x=datos["X"], y=datos["Y"], c=datos["Z"], label="Puntos")
plt.axis('equal')
plt.show()
```

```python
# Realizamos un histograma para cada clase
sns.FacetGrid( datos, hue="Z", height=5, aspect=3, palette=["#D62728","#1F77B4"] ).map(plt.hist,"X", alpha=0.5)
plt.show()
```

```python
matriz_diseño = datos.values[:,0:2]
y_real        = datos.values[:,  2]

modelo_lda = LinearDiscriminantAnalysis(n_components=None).fit( matriz_diseño, y_real )
modelo_lda
```

```python
y_estimada = modelo_lda.predict( matriz_diseño )
y_estimada[0:10]
```

```python
modelo_lda.predict_proba( matriz_diseño )
```

```python
matriz_confusion = confusion_matrix( y_real, y_estimada )
matriz_confusion
```

```python
print( "Precisión: ", ( matriz_confusion[0,0] + matriz_confusion[1,1] ) / np.sum( matriz_confusion ) )
print( "Precisión: ", accuracy_score( y_true = y_real, y_pred = y_estimada ) )
```

## Datos :: Iris

```python
# Leemos nuestros archivos 
datos = pd.read_csv("iris.csv")   # Datos con 3 clases
#datos = pd.read_csv("rock.csv")   # Datos con 12 clases
datos.sample(10)
```

```python
datos["Species"].unique()
```

```python
sns.pairplot(datos, hue = "Species", height=2, aspect=2)
plt.plot()
```

```python
matriz_diseño = datos.iloc[:,0:4]
y_real        = datos.iloc[:,  4]
matriz_diseño.head()
```

```python
# Generamos nuestro modelo
modelo_lda = LinearDiscriminantAnalysis(n_components=None).fit( matriz_diseño, y_real )
modelo_lda
```

```python
y_estimada = modelo_lda.predict( matriz_diseño )
y_estimada[0:10]
```

```python
modelo_lda.predict_proba( matriz_diseño.head() )
```

```python
matriz_confusion = confusion_matrix( y_real, y_estimada )
matriz_confusion
```

```python
print( "Precisión: ", accuracy_score( y_true = y_real, y_pred = y_estimada ) )
```

```python
modelo_lda = LinearDiscriminantAnalysis(n_components=None)
k_fold     = KFold(n_splits=10, shuffle=True, random_state=0)
score      = cross_val_score( modelo_lda, matriz_diseño, y_real, cv=k_fold, n_jobs=1, scoring="accuracy")
print( "Precisiones: ",       score )
print( "Precisión puntual: ", np.mean(score) )
```
